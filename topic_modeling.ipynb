```python
df = df_cleaned_final.copy()
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    """
    Cleans text by:
    - Removing special characters and numbers.
    - Converting to lowercase.
    - Removing stopwords.
    - Lemmatizing the words to their base form.
    Returns a single string representing the cleaned text.
    """

    # Remove special characters and numbers
    text = re.sub(r'[^A-Za-z\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove stopwords and lemmatize
    words = text.split()
    cleaned_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    # Return the cleaned text as a single string
    return ' '.join(cleaned_words)
```

```python
# Apply to 'text' and 'title' column
df['text_cleaned'] = df['text'].apply(clean_text)
df['title_cleaned'] = df['title'].apply(clean_text)
```

```python
# Combine title and text to create a new column called "title_text_cleaned"
df['title_text_cleaned'] = df['text_cleaned'] + " " + df['title_cleaned']
```

```python
# Turn the column 'title_text_cleaned' to a list for further analysis
# A list of documents (strings). The topic model will be trained on these documents.
texts = df['title_text_cleaned'].tolist()
```

### Start Topic Modeling with 12 topics

A topic is not a single word but rather a distribution of words.

For example:

Topic 1 might consist of words like {"economy": 0.2, "trade": 0.1, "inflation": 0.05, ...}.

Topic 2 might include words like {"technology": 0.3, "innovation": 0.1, "AI": 0.05, ...}.

When a document is analyzed, the topic model determines the proportion of the document that belongs to each topic.

```python
%%time
tm = ktrain.text.get_topic_model(
    texts=texts,
    n_topics=12,
    n_features=10000, # Limits the vocabulary size to the top 10,000 most frequent terms in the dataset.
    min_df=0.001, # Words that appear in less than 0.1% of the documents are ignored.
    max_df=0.5, # Words that appear in more than 50% of the documents are ignored.
    stop_words='english', # Common English stop words (like "the," "and," "is") are removed before training to focus on meaningful terms.
    model_type='lda',
    lda_max_iter=5, # The model will iterate up to 5 times to converge on a solution for topic distributions.
    verbose=1) # Enables logging to show progress and information during training. It helps monitor the training process.
```

```python
tm.print_topics()
```

```python
# For each document in the dataset (texts), it calculates the topic probabilities using the trained model.
# If the most probable topic for a document has a probability equal to or greater than 0.25, that topic is assigned to the document.
# Purpose: The build method assigns topics to the input texts (documents) based on the trained topic model (tm).
# For each document in the dataset (texts), it calculates the topic probabilities using the trained model.
# Assigns topics to documents based on the threshold parameter.

%%time
tm.build(texts, threshold=0.25)
```

```python
tm.print_topics(show_counts=True)
```

Visualize the topic distributions for documents after training a topic model (tm) in ktrain

```python
# Retrieves the topic distributions for all documents in the dataset
# A matrix where each row corresponds to a document and each column corresponds to a topic. The values are the probabilities of each topic for the document.
doc_topics=tm.get_doctopics()
```

```python
# Visualize the documents in a 2D space based on their topic distributions.
# Reduces the dimensionality of the doc_topics matrix (e.g., using t-SNE or UMAP) to project documents into 2D.
# Each document is represented as a point in this 2D space, with colors or labels indicating the dominant topic for each document.
# Clusters of Documents: Documents with similar topic distributions will be closer to each other.
# Dominant Topics: Points may be color-coded by the most probable topic, helping you identify clusters for each topic.
tm.visualize_documents(doc_topics=tm.get_doctopics())
```

### Assigns topics to documents based on the topic probabilities predicted by the topic model (tm)

```python
# Generates the topic probability distribution for each document in texts using the trained topic model (tm).
# A 2D array (topic_probs) where each row corresponds to a document and each column corresponds to a topic.
# Each value is the probability of the document belonging to that topic.
topic_probs = tm.predict(texts)

# Specifies the minimum probability a topic must have to be assigned to a document.
# Only topics with probabilities ≥ 0.25 will be considered.
threshold = 0.25

# An empty list to store the assigned topic for each document.
topic_assignments = []

# For each document, select the most probable topic and check if it exceeds the threshold.
# Loops through each document's topic probabilities and assigns a topic based on the threshold.
for probs in topic_probs:
    top_topic = np.argmax(probs)  # Select the most probable topic
    if probs[top_topic] >= threshold:  # If the probability of that topic is above the threshold
        topic_assignments.append(top_topic)  # Assign the topic
    else:
        topic_assignments.append(None)  # Assign None if below threshold

# Add a new column to the dataframe containing the predicted topic indices
df['topic'] = topic_assignments
```

```python
# Replace NaN values in the 'topic' column with 'other'
df['topic'] = df['topic'].fillna('other')

# Convert all values in the 'topic' column to strings
df['topic'] = df['topic'].astype(str)
```

Documents with similar topic probabilities will appear closer in 2D.

While relative positions are preserved, the 2D representation is an approximation and may not perfectly reflect distances in the original topic space.
